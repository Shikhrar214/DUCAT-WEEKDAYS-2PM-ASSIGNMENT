## String len() function, indexing, error

```python
name = "Shikhar"
full_name = "Shikhar Srivastav."
print(len(name))
print(len(full_name))
print(name[0])
print(name[1])
# print(name[8]) #(indexError)string index out of range
# print("0o = ", name[03]) #(SyntaxError)leading zeros in decimal integer literals are not permitted; use an 0o prefix for octal integers
print("0o = ", name[0o3])
print(name[-1])
print(name[-2])
print("0o = ", name[-0o3])
# slicing
# print(name[start:end:step])
print("name === ",name[0:7:1])
print("name === ",name[-7:-1:1])
print("name === ",name[::-1])
print("name === ",name[-1::1])
```

## Changing Case conversion method in python [upper(),lower(),capitalize(),title(),swapcase()]

```python
print("01. ", "My name is Shikhar".upper())
print("02. ","My NAME is Shikhar".lower())
print("03. ","My name is Shikhar. i live in noida.".capitalize())
print("04. ","My name is Shikhar".casefold())
print("05. ","shikhar srivastav".title())
print("06. ","My name is Shikhar".swapcase()) #print upper case to lowercase and lower case to upper case.
```

## Searching method [find(), rfind(), index(), rindex(), count(), startwith(), endswith()]

```python
text = "My name is shikhar"

print(text.find("S"))
print(text.find("s"))
print(text.rfind("S"))
print(text.rfind("s"))
# print(text.index("S")) #substring not found
print("index", text.index("s"))
print("rindex", text.rindex("s"))
print("count", text.count("a"))
print("count", text.count("z"))
print(text.startswith("h"))
print(text.startswith("M"))
print(text.endswith("r"))
print(text.endswith("M"))
```

## Validation method. [isalpha(),idigit(),isalnum(),isspace(),islower(),isupper()]

```python
print("01. ","1234125567".isdigit())
print("02. ","shikhar".isalpha())
print("03. ","90abc123".isalnum())
print("04. "," ".isspace())
print("05. ","shikhar".islower())
print("06. ","SHIKHAR".isupper())
print("07. ","hello Shikhar".isascii())
print("08. ","909".isdecimal())
print("09. ","class".isidentifier())
print("10. ","13242".isnumeric())
# print("11. ","abc#12\n".isprintable()) #(\n, \t) is not printable
print("11. ","abc#12".isprintable())
print("12. ","Shikhar Srivastav".istitle())
# "123".isdecimal()      # True
# "12.3".isdecimal()     # False
# "-123".isdecimal()    # False
# "123 ".isdecimal()     #false
```

## Splitting and Joining [split(), rsplit(), splitlines(), join(), partition(), rpartition()]

```python
print("split = ","a,b,c,d,e,f".split(",",2))
print("rsplit = ","a-b-c-d-e-f".rsplit("-",-1))
print("rsplit = ","a-b-c-d-e-f".rsplit("-",1))
print("splitlines = ","my name is shikhar \n currently I am living in noida.".splitlines())
# print("join = ", "-".join([2000, 12,03])) #SyntaxError: leading zeros in decimal integer literals are not permitted; use an 0o prefix for octal integers
print("join = ", "-".join(["2000", "12","03"]))
print("key=value".partition("="))
print("file.name.txt".rpartition("."))
```

## Tirmming method [replace(), strip(), lstrip(), rstrip(), removeprefix(), removesuffix()]

```python
text = "  hello  "
print(text.strip()) 
print(text.lstrip())
print(text.rstrip())
print("mr. Shikhar".removeprefix("mr."))
print("Shikhar Srivastav ji".removesuffix("ji"))
```

## Formatting and alignment Method [format(), f-string()]

```python
name = "Shikhar"
print(f"My name is {name}")
txt = "For only {price:.2f} dollars!"
print(txt.format(price = 49))
print("Hello {}".format("Shikhar"))
data = {"name": "Shikhar"}
print("Hello {name}".format_map(data))
print("Hi".center(10, "-"))
print("Hi".ljust(10), "Shikhar")
print("Hi".rjust(10))
print("42".zfill(5))
```

## Replace and translate method [replace(), translate(), maketrans())]

```python
print("I love Java".replace("Java", "Python"))
table = str.maketrans("ae", "12")
print("apple".translate(table))
table = str.maketrans({"a": "1"})
print(table)
```
